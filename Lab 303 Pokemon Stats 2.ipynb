```python
pokemon['Combo Type'] = pokemon['Type 1'] + ('-'+pokemon['Type 2'].fillna(' '))
pokemon.head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
type1 = pokemon[['Type 1']]
print(type1)
encoder = OneHotEncoder().fit(type1)
encoded = encoder.transform(type1).toarray()
cols = encoder.get_feature_names_out()
type1_encoded = pd.DataFrame(encoded,columns=cols)
type1_encoded
```

```python
type2 = pokemon[['Type 2']]
encoder = OneHotEncoder().fit(type2)
encoded = encoder.transform(type2).toarray()
cols2 = encoder.get_feature_names_out()
type2_encoded = pd.DataFrame(encoded,columns=cols2)
type2_encoded = type2_encoded
type2_encoded
```

```python
types_encoded = type1_encoded+type2_encoded
types_encoded
```
