```python
pokemon.Name = pokemon.Name.apply(lambda x: re.sub(r'(.+)(Mega.+)',r'\2', x))
pokemon.Name = pokemon.Name.apply(lambda x: re.sub(r'(.+)(Prima1.+)',r'\2', x))
pokemon.Name = pokemon.Name.apply(lambda x: re.sub(r'(HoopaHoopa)(.+)','Hoopa'+r'\2', x))
```

```python
pokemon
```

```python
pokemon['A/D Ratio'] = pokemon['Attack']/pokemon['Defense']
pokemon
```

```python
pokemon_ad = pokemon['A/D Ratio'].idxmax()
pokemon_max_ad = pokemon.loc[pokemon_ad]
print(pokemon_max_ad)
```

```python
pokemon_ad2 = pokemon['A/D Ratio'].idxmin()
pokemon_min_ad = pokemon.loc[pokemon_ad2]
print(pokemon_min_ad)
```

```python
pokemon['Type 2'] = pokemon['Type 2'].fillna('')
pokemon['Combo Type'] = pokemon['Type 1'] + (' ' + pokemon['Type 2']).fillna(' ')
pokemon
```

```python
pokemon_top_ratio = pokemon.nlargest(5, 'A/D Ratio')
pokemon_top_ratio
```

```python
pokemon_avg = pokemon.groupby('Combo Type').mean(numeric_only=True).reset_index()
pokemon_avg[pokemon_avg['Combo Type'].isin(pokemon_top_ratio)]
pokemon_avg
```
